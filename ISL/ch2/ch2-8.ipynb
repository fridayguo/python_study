（c) 4、 创建一个新的定性变量，名为 Elite ，合并Top10perrc变量，是否有超过 50% 以上的学生来自排名在前 10% 顶尖高中的情况将大学分成两组 O

```python
df['Elite'] = ''
df.loc[df.Top10perc>=50,'Elite'] = 'Yes'
df.loc[df.Top10perc<50,'Elite'] = 'No'

df['Elite'].value_counts()
```

```python
sns.boxplot(df['Elite'], df['Outstate'])
plt.show()
```
